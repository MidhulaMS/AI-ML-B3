```python
# Weather Data Analyzer


def average_temperature(temps):
    return sum(temps) / len(temps)

def average_humidity(humidities):
    return sum(humidities) / len(humidities)

def main(temperature, humidity):
    if temperature > 30 and humidity < 60:
        return "Sunny"
    elif humidity >= 80:
        return "Rainy"
    elif 20 <= temperature <= 30:
        return "Cloudy"
    else:
        return "Cold"


temperatures = [24, 32, 30, 32, 27]
humidities = [35, 45, 45, 50, 20]


avg_temp = average_temperature(temperatures)
avg_hum = average_humidity(humidities)


print(f"Weather Analysis \n")

for i in range(len(temperatures)):
    condition = main(temperatures[i], humidities[i])
    print(f"Day {i+1}: Temperature = {temperatures[i]}°C, "f"Humidity = {humidities[i]}%, "f"Condition = {condition}")

print("\nAverage Temperature:", round(avg_temp, 2), "°C")
print("Average Humidity:", round(avg_hum, 2), "%")
```
